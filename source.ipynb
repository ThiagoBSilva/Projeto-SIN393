```
#Importando bibliotecas e pacotes necessários para construir o modelo de CNN
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```
#Criando uma classe Net, o modelo da nossa Rede Neural

class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 32, 5) #input é uma imagem, output são 32 convoluções, kernel/janela de 5x5
        self.conv2 = nn.Conv2d(32, 64, 5) #input é o output da camada anterior, output são 64 convoluções...
        self.conv3 = nn.Conv2d(64, 128, 5)
        
        #É necessário passar uma imagem aplainada (flattened) para as layers fully-connected
        #Para isso criamos uma função para poder realizar esse processo
        x = torch.randn(50, 50).view(-1, 1, 50, 50)
        self._to_linear = None
        self.convs(x)
        
        self.fc1 = nn.Linear(self._to_linear, 512) #Aplainamento
        self.fc2 = nn.Linear(512, 26) #512 de entrada e 26 de saída por causa das classes (A, B, C...)
        
    def convs(self, x):
        #Max Pooling 2x2
        x = F.max_pool2d(F.relu(self.conv1(x)), (2, 2))
        x = F.max_pool2d(F.relu(self.conv2(x)), (2, 2))
        x = F.max_pool2d(F.relu(self.conv3(x)), (2, 2))
        
        if self._to_linear is None:
            self._to_linear = x[0].shape[0]*x[0].shape[1]*x[0].shape[2]
        return x
    
    #Função que controla como os dados serão mandados para "frente"
    def forward(self, x):
        x = self.convs(x)
        x = x.view(-1, self._to_linear) #Aplaina X antes
        x = F.relu(self.fc1(x))
        x = self.fc2(x) #Camada de saída, sem ativação Relu aqui
        return F.softmax(x, dim = 1)
    
    
net = Net()
print(net)      
```

```
#Precisamos separar os nossos dados em X (features) e y (labels)
X = torch.Tensor([i[0] for i in training_data]).view(-1, 50, 50)
X = X/255.0
y = torch.Tensor([i[1] for i in training_data])
```

```
VAL_PCT = 0.1 #Reservando 10% dos nossos dados para validação do modelo
val_size = int(len(X)*VAL_PCT)
print(val_size)
```

```
#Definindo os X (features) e y(labels) de teste e treino
train_X = X[: -val_size]
train_y = y[: -val_size]

test_X = X[-val_size: ]
test_y = y[-val_size: ]

print(len(train_X), len(test_X))
```

```
#Agora precisamos treinar o nosso modelo

import torch.optim as optim
#Para minimizar a taxa de perda do modelo, devemos usar um otimizador
optimizer = optim.Adam(net.parameters(), lr = 0.001) #Otimizador Adam para otimizar os parâmetros que variam, passo de aprendizado de 0.001
loss_function = nn.MSELoss()
```

```
def fwd_pass(X, y, train=False):

  if train:
    #Zera os gradientes ao fim de um batch
    net.zero_grad()
  outputs = net(X)
    
  #Conta quantas combinações foram feitas no processo de treino por época   
  #Depois calcula a acurácia por amostra
  matches = [torch.argmax(i)==torch.argmax(j) for i, j in zip(outputs, y)]
  acc = matches.count(True)/len(matches)
  loss = loss_function(outputs, y)

  if train:
    loss.backward()

    #Atualiza o otimizador
    optimizer.step()

  return acc, loss
```

```
def test(size=96):
  X, y = test_X[:size], test_y[:size]
  val_acc, val_loss = fwd_pass(X.view(-1, 1, 50, 50), y)
  return val_acc, val_loss
```

```
import time

MODEL_NAME = f"model-{int(time.time())}"

#Definindo a função de treino
def train(net):

    #Nós precisamos iterar sobre os nossos dados. Para isso decidimos qual o tamanho dos batches usar.
    BATCH_SIZE = 1000
    #EPOCHS é o número de vezes que será iterado sobre todos os dados do conjunto
    EPOCHS = 15

    with open("model.log", "a") as f:
       for epoch in range(EPOCHS):
          for i in tqdm(range(0, len(train_X), BATCH_SIZE)):
              batch_X = train_X[i:i+BATCH_SIZE].view(-1, 1, 50, 50)
              batch_y = train_y[i:i+BATCH_SIZE]

              acc, loss = fwd_pass(batch_X, batch_y, train=True)

              #print(f"Epoch: {epoch}. Perda: {round(float(loss),4)}". "Acc: {round(float(acc),2)}")
              #f.write(f"{MODEL_NAME}, {round(time.time(), 3)}, in_sample, {round(float(acc),2)}, {round(float(loss),4)}\n")

              #A cada 30 passos de iteração, será calculado a acc de validação e a perda(loss) de validação
              #Os resultados serão escritos num arquivo chamado "model.log" que haverá o registro de tempo, acc e perda in-sample e acc e perda de validação.
              if i % 1 == 0:
                  val_acc, val_loss = test(size=BATCH_SIZE)
                  f.write(f"{MODEL_NAME},{round(time.time(),3)},{round(float(acc),2)},{round(float(loss), 4)},{round(float(val_acc),2)},{round(float(val_loss),4)},{epoch}\n")

          print(f"Epoch: {epoch}, Acc: {round(float(acc),2)}, Acc_val: {round(float(val_acc),2)}, Loss: {round(float(loss),4)}, Loss_val: {round(float(val_loss),4)}")  

train(net)
```

```
import matplotlib.pyplot as plt
from matplotlib import style

style.use("ggplot")

model_name = MODEL_NAME

def accxloss_graph(model_name):
  contents = open("model.log", "r").read().split("\n")

  epochs = []
  times = []
  accuracies = []
  losses = []

  val_accs = []
  val_losses = []

  for c in contents:
    if model_name in c:
      name, timestamp, acc, loss, val_acc, val_loss, epoch = c.split(",")

      epochs.append(int(epoch))
      times.append(float(timestamp))
      accuracies.append(float(acc))
      losses.append(float(loss))

      val_accs.append(float(val_acc))
      val_losses.append(float(val_loss))

  fig = plt.figure()

  ax1 = plt.subplot2grid((2,1), (0,0))
  ax2 = plt.subplot2grid((2,1), (1,0), sharex = ax1)

  ax1.plot(times, accuracies, label = "acc_in_sample")
  ax1.plot(times, val_accs, label = "acc_valid")
  ax1.legend(loc= 2)
  ax2.plot(times, losses, label = "loss_in_sample")
  ax2.plot(times, val_losses, label = "loss_valid")
  ax2.legend(loc=2)
  plt.show()

accxloss_graph(model_name)
```

```
#Definindo a função de teste

predict = []
real = []

def test_val(net):
    correct = 0
    total = 0
    with torch.no_grad():
        for i in tqdm(range(len(test_X))):
            real_class = torch.argmax(test_y[i])
            net_out = net(test_X[i].view(-1, 1, 50, 50))[0] #Retorna uma lista que é o output dado pelo modelo na validação
            predicted_class = torch.argmax(net_out)
        
            predict.append(predicted_class.numpy())
            real.append(real_class.numpy())

            if predicted_class == real_class:
                correct += 1
            total += 1
    print("Acurácia: ", round(correct/total, 3))

test_val(net)
```

```
real_int = []
predict_int = []

for a in real:
  real_int.append(a.item())

for a in predict:
  predict_int.append(a.item())

print(real_int)
print(predict_int)
```
